```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url_link = "https://en.wikipedia.org/wiki/states_and_union_territories_of_India"

# Önce kimliğini ve kurallarını belirle.
headers = {
    'User-Agent': 'MyCoolBot/1.0 (https://example.com/mycoolbot; myemail@example.com)'
}

# Sonra, bu kimliği kullanarak isteği gönder.
# headers parametresini get() fonksiyonunun içine ekliyoruz.
result = req.get(url_link, headers=headers) 

# Gelen içeriği (HTML) bir değişkene ata.
html_content = result.text
print(html_content)
```

```python
doc = BeautifulSoup(html_content, "html.parser")
doc
```

```python
article_titles = doc.find_all('h1')
for title in article_titles:
    print(title.text)
```
